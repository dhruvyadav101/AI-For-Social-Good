```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Distribution of classes
plt.figure(figsize=(8, 6))
sns.countplot(x='label', data=df)
plt.title('Distribution of Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Convert 'post_text' column to string type
df['post_text'] = df['post_text'].astype(str)

# Combine all tweets into one string
all_words = ' '.join([text for text in df['post_text']])

# Create and display the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_words)

plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Processed Tweets')
plt.show()
```

```python
from collections import Counter

# Split words and count frequencies
words = all_words.split()
word_freq = Counter(words)

# Convert to DataFrame for plotting
word_freq_df = pd.DataFrame(word_freq.items(), columns=['word', 'freq']).sort_values(by='freq', ascending=False).head(20)

# Plot the frequencies
plt.figure(figsize=(12, 8))
sns.barplot(x='freq', y='word', data=word_freq_df)
plt.title('Top 20 Most Frequent Words in Processed Tweets')
plt.xlabel('Frequency')
plt.ylabel('Word')
plt.show()
```
